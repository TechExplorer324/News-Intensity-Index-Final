```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import datetime as dt  
```

```python
'''vals = pd.read_excel("Gemini.xlsx")
vals1 = pd.read_excel("ChatGPT.xlsx")
vals2 = pd.read_excel("DeepSeek.xlsx")'''
#vals = pd.read_excel("ChatGPT.xlsx")

vals = pd.read_excel("GeminiC.xlsx")
vals1 = pd.read_excel("ChatGPTC.xlsx")
vals2 = pd.read_excel("DeepSeekC.xlsx")

print(vals2)
```

```python
vals['Day'] = pd.to_datetime(vals['Day'], format = '%d-%m-%Y')
vals.rename(columns={'Day':'Date'}, inplace=True)

vals1['Day'] = pd.to_datetime(vals1['Day'], format = '%d-%m-%Y')
vals1.rename(columns={'Day':'Date'}, inplace=True)

vals2['Day'] = pd.to_datetime(vals2['Day'], format = '%d-%m-%Y')
vals2.rename(columns={'Day':'Date'}, inplace=True)


print(vals1)
```

```python
#tag = 'Rolling Change'
tag = 'Change %'
```

```python
#prices = pd.read_csv("GOOGL Historical Data.csv")
prices = pd.read_csv("Nifty IT Historical Data.csv")
print(prices)
```

```python
prices['Date'] = pd.to_datetime(prices['Date'], format = '%d-%m-%Y')
print(prices)
```

```python
merged_df = pd.merge(vals[['Date','Gemini: (Worldwide)']], prices[['Date',tag]], on='Date', how='inner')
merged_df = pd.merge(merged_df, vals1[['Date','ChatGPT: (Worldwide)']], on='Date', how='inner')
merged_df = pd.merge(merged_df, vals2[['Date','DeepSeek: (Worldwide)']], on='Date', how='inner')
print("\n Merged DataFrame:\n", merged_df)
```

```python
'''merged_df['Rolling Change'] = merged_df['Rolling Change'].shift(-1)
merged_df.dropna(inplace=True)

merged_df['Rolling Change'] = merged_df['Rolling Change'].astype(str).str.replace('%','').astype(float)
print(merged_df)'''

merged_df[tag] = merged_df[tag].shift(-1)
merged_df.dropna(inplace=True)

merged_df[tag] = merged_df[tag].astype(str).str.replace('%','').astype(float)
print(merged_df)
```

```python
merged_df['Numerical_date'] = merged_df['Date'].map(dt.datetime.toordinal)
```

```python
#x = merged_df[['Numerical_date','Gemini: (Worldwide)']]
x = merged_df[['Gemini: (Worldwide)', 'ChatGPT: (Worldwide)', 'DeepSeek: (Worldwide)']]
y = merged_df[tag]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
```

```python
model1 = LinearRegression()
model1.fit(x_train, y_train )
y_pred1 = model1.predict(x_test)
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred1)
r2 = r2_score(y_test, y_pred1)
```

```python
print("Linear Regression Predictions:\n")
print("Mean Squared Error:\n", mse)
print("R^2 Score:\n", r2)
```

```python
print(f"Model Intercept: {model1.intercept_:.2f}")
print(f"Model Coefficients (Gemini Search Volume): {model1.coef_}")
```

Lets Try Light BGM

```python
merged_df1 = merged_df.rename(columns={'Gemini: (Worldwide)': 'Gemini_Worldwide'})

df = merged_df1
x = merged_df[['Gemini: (Worldwide)', 'ChatGPT: (Worldwide)', 'DeepSeek: (Worldwide)']]
y = merged_df1[tag]

x1 = x

x1 = x1.rename(columns={
    'Gemini: (Worldwide)': 'Gemini',
    'ChatGPT: (Worldwide)': 'ChatGPT',
    'DeepSeek: (Worldwide)': 'DeepSeek'
})

x1_train, x1_test, y_train, y_test = train_test_split(x1, y, test_size = 0.2, random_state = 42)

lgb_train = lgb.Dataset(x1_train, y_train)
lgb_eval = lgb.Dataset(x1_test, y_test, reference=lgb_train)

lgbm_params = {
    'objective': 'regression',
    'metric': 'rmse',
    'boosting_type': 'gbdt',
    'num_leaves': 31,
    'learning_rate': 0.05,
    'feature_fraction': 0.9,
    'seed': 42
}
lgb_model = lgb.train(
    lgbm_params,
    lgb_train,
    num_boost_round=100,
    valid_sets=lgb_eval,
    callbacks=[lgb.callback.early_stopping(10, verbose=False)] # Stop early if no improvement
)
lgb_pred = lgb_model.predict(x_test, num_iteration=lgb_model.best_iteration)
```

```python
mse = mean_squared_error(y_test, lgb_pred)
r2 = r2_score(y_test, lgb_pred)

print("LGB:\n")
print("Mean Squared Error:\n", mse)
print("R^2 Score:\n", r2)
```

```python
'''scalar = StandardScaler()
df = merged_df
x = scalar.fit_transform(merged_df[['Gemini: (Worldwide)', 'ChatGPT: (Worldwide)', 'DeepSeek: (Worldwide)']])
y = scalar.fit_transform(merged_df[['Rolling Change']])
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
model = SVR(kernel='rbf', C=10, gamma=0.1)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)'''

scalar_x = StandardScaler()
scalar_y = StandardScaler()

x = merged_df[['Gemini: (Worldwide)', 'ChatGPT: (Worldwide)', 'DeepSeek: (Worldwide)']]
y = merged_df[tag]

# Scale X
x_scaled = scalar_x.fit_transform(x)

# Scale Y and immediately flatten it to 1D for SVR/MLP
y_scaled = scalar_y.fit_transform(y.values.reshape(-1, 1)).ravel() # Reshape and then flatten

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size = 0.2, random_state = 42)

model = SVR(kernel='rbf', C=10, gamma=0.1)
model.fit(x_train, y_train)
y_pred_scaled = model.predict(x_test)
```

```python
'''mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("SVR Predictions:\n")
print("Mean Squared Error:\n", mse)
print("R^2 Score:\n", r2)'''

# Inverse transform to get predictions and actual values back to original scale
y_pred = scalar_y.inverse_transform(y_pred_scaled.reshape(-1, 1))
y_test_original = scalar_y.inverse_transform(y_test.reshape(-1, 1))

# Calculate metrics on the original, unscaled values
mse = mean_squared_error(y_test_original, y_pred)
r2 = r2_score(y_test_original, y_pred)

print("SVR Predictions (Metrics on original scale):\n")
print("Mean Squared Error:\n", mse)
print("R^2 Score:\n", r2)
```
