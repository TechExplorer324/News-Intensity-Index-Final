```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import datetime as dt  
```

```python
vals = pd.read_excel("Gemini.xlsx")
#vals = pd.read_excel("ChatGPT.xlsx")
print(vals)
```

```python
vals['Day'] = pd.to_datetime(vals['Day'], format = '%d-%m-%Y')
vals.rename(columns={'Day':'Date'}, inplace=True)
print(vals)
```

```python
prices = pd.read_csv("GOOGL Historical Data.csv")
print(prices)
```

```python
prices['Date'] = pd.to_datetime(prices['Date'], format = '%d-%m-%Y')
print(prices)
```

```python
merged_df = pd.merge(vals[['Date','Gemini: (Worldwide)']], prices[['Date','Price']], on='Date', how='inner')
print("\n Merged DataFrame:\n", merged_df)
```

```python
merged_df['Numerical_date'] = merged_df['Date'].map(dt.datetime.toordinal)
```

```python
#x = merged_df[['Numerical_date','Gemini: (Worldwide)']]
x = merged_df[['Gemini: (Worldwide)']]
y = merged_df['Price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
```

```python
model1 = LinearRegression()
```

```python
model1.fit(x_train, y_train )
```

```python
y_pred1 = model1.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred1)
r2 = r2_score(y_test, y_pred1)
```

```python
print("Linear Regression Predictions:\n")
print("Mean Squared Error:\n", mse)
print("R^2 Score:\n", r2)
```

```python
print(f"Model Intercept: {model1.intercept_:.2f}")
print(f"Model Coefficients (Gemini Search Volume): {model1.coef_}")
```

Lets Try Light BGM

```python
merged_df1 = merged_df.rename(columns={'Gemini: (Worldwide)': 'Gemini_Worldwide'})

df = merged_df1
x = merged_df1[['Gemini_Worldwide']]
y = merged_df1['Price']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

lgb_train = lgb.Dataset(x_train, y_train)
lgb_eval = lgb.Dataset(x_test, y_test, reference=lgb_train)

lgbm_params = {
    'objective': 'regression',
    'metric': 'rmse',
    'boosting_type': 'gbdt',
    'num_leaves': 31,
    'learning_rate': 0.05,
    'feature_fraction': 0.9,
    'seed': 42
}
lgb_model = lgb.train(
    lgbm_params,
    lgb_train,
    num_boost_round=100,
    valid_sets=lgb_eval,
    callbacks=[lgb.callback.early_stopping(10, verbose=False)] # Stop early if no improvement
)
lgb_pred = lgb_model.predict(x_test, num_iteration=lgb_model.best_iteration)
```

```python
mse = mean_squared_error(y_test, lgb_pred)
r2 = r2_score(y_test, lgb_pred)

print("LGB:\n")
print("Mean Squared Error:\n", mse)
print("R^2 Score:\n", r2)
```

NOW LETS TRY SVR - SUPPORT VECTOR REGRESSOR

```python
from sklearn.svm import SVR # For Support Vector Regression
from sklearn.neighbors import KNeighborsRegressor # For KNN Regression
from sklearn.neural_network import MLPRegressor # For Neural Network (MLP)
```

```python
scalar = StandardScaler()
df = merged_df
x = scalar.fit_transform(merged_df[['Gemini: (Worldwide)']])
y = scalar.fit_transform(merged_df[['Price']])
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
model = SVR(kernel='rbf', C=10, gamma=0.1)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("SVR Predictions:\n")
print("Mean Squared Error:\n", mse)
print("R^2 Score:\n", r2)
```
